````python
import json

def compress_jsonl(input_file, output_file):
    with open(input_file, 'r') as infile:
        current_json = ""
        for line in infile:
            current_json += line
            try:
                json_obj = json.loads(current_json)
                with open(output_file, 'a') as outfile:
                    outfile.write(json.dumps(json_obj, separators=(',',':')) + '\n')
                current_json = ""
            except json.JSONDecodeError:
                continue
    
compress_jsonl('courts_for_gpt.jsonl', 'compressed_courts_for_comparing.jsonl')
````

````python
import re

df['id'] = df['id'].astype(str)
df_xslx['Case ID'] = df_xslx['Case ID'].astype(str)

def extract_ids(case_id):
    matches = re.findall(r'\d+', case_id)
    return matches  

decision_dict = {}
for _, row in df_xslx.iterrows():
    case_ids = extract_ids(row['Case ID']) 
    for case_id in case_ids:
        decision_dict[case_id] = row['Decision For President']

df['Decision For President'] = df['id'].map(decision_dict).fillna(np.nan)

df.head()
````

````python
import json

def create_batch_4o_mini(custom_id, method, url, messages, output_file):
    batch_entry = {
        "custom_id": custom_id,
        "method": method,
        "url": url,
        "body": {
            "model": "gpt-4o-mini",
            "messages": messages,
            "max_tokens": 9000
        }
    }
    with open(output_file, 'a') as outfile:
        outfile.write(json.dumps(batch_entry, separators=(',',':')) + '\n')

def create_batch_4o(custom_id, method, url, messages, output_file):
    batch_entry = {
        "custom_id": custom_id,
        "method": method,
        "url": url,
        "body": {
            "model": "gpt-4o",
            "messages": messages,
            "max_tokens": 9000
        }
    }
    with open(output_file, 'a') as outfile:
        outfile.write(json.dumps(batch_entry, separators=(',',':')) + '\n')
        
output_file_4o_mini = 'batch_input_4o_mini.jsonl'
output_file_4o = 'batch_input_4o.jsonl'
method = "POST"
url = "/v1/chat/completions"

line_count = 0
for index, row in df_sorted.iterrows():
    if line_count >= 100:  # Limit to first 100 lines
        break
    valid_column = next(
        (col for col in ['html', 'xml_harvard', 'plain_text', 'html_lawbox'] if pd.notnull(row[col]) and row[col] != ""), 
        None
    )

    custom_id = f"request-{index + 1}"
    messages = [
    {
        "role": "system", 
        "content": "Your goal is to determine whether each opinion is in favor of or opposing the President. You will receive the text of a Supreme Court opinion, and you should analyze the sentiment and key phrases to make your determination."
    },
    {
        "role": "user", 
        "content": f"""Here is the text of a Supreme Court opinion: {row[valid_column]}.
        Based on this opinion, please analyze the sentiment and determine if it is in favor of or opposing the President. 
        Return your response in JSON format:
            {{
                "resource": "{row['resource_uri']}",
                "opinion": "with values 'in favor' or 'opposed'",
                "explanation": "a brief explanation of your conclusion"
            }}
        """
    }
]

    create_batch_4o_mini(custom_id, method, url, messages, output_file_4o_mini)
    create_batch_4o(custom_id, method, url, messages, output_file_4o)
    line_count += 1  # Increment line count

````

````python
import os

def test(input_file, output_files, num):
    with open(input_file, 'r') as infile:
        line_count = 0 
        for line in infile:
            if line_count >= 100: 
                break
            try:
                json_obj = json.loads(line)
                output_index = line_count // num
                os.makedirs(os.path.dirname(output_files[output_index]), exist_ok=True)
                with open(output_files[output_index], 'a') as outfile:
                    outfile.write(json.dumps(json_obj) + '\n')
            except json.JSONDecodeError:
                continue
            line_count += 1  

mini_files = [f'4o_mini/{i}_mini.jsonl' for i in range(1, 2)]
test('batch_input_4o_mini.jsonl', mini_files, 100)

full_files = [f'4o/{i}.jsonl' for i in range(1, 11)]
test('batch_input_4o.jsonl', full_files, 10)
````

````python
import time
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI, RateLimitError, APIError
client = OpenAI()

def cancel_all_batches():
    try:
        batches = client.batches.list()
        for batch in batches:
            cancellable_statuses = ["queued", "in_progress"]
            if batch.status in cancellable_statuses:
                client.batches.cancel(batch.id)
            wait_for_cancel(batch.id)
    except APIError as e:
        print(f"API Error while cancelling batches: {str(e)}")
    except Exception as ex:
        print(f"Unexpected error while cancelling batches: {str(ex)}")

def wait_for_cancel(batch_id, timeout=60):
    start_time = time.time()
    while time.time() - start_time < timeout:
        try:
            batch = client.batches.retrieve(batch_id)
            if batch.status == "cancelled":
                print(f"Batch {batch_id} successfully canceled.")
                return
            elif batch.status in ["failed", "completed"]:
                print(f"Batch {batch_id} has status {batch.status}.")
                return
            else:
                print(f"Waiting for batch {batch_id} to cancel. Current status: {batch.status}")
        except APIError as e:
            print(f"API Error while checking batch status: {str(e)}")
            break
        time.sleep(5)
    print(f"Timeout reached while waiting for batch {batch_id} to cancel.")

cancel_all_batches()
````

````python
def process_files(files_to_process, file_ids, batch_ids, pause = 10):
    for file in files_to_process:
        try:
          batch_input_file = client.files.create(
            file=open(file, "rb"),
            purpose="batch"
          )

          batch_input_file_id = batch_input_file.id
          file_ids.append(batch_input_file_id)

          batch = client.batches.create(
              input_file_id=batch_input_file_id,
              endpoint="/v1/chat/completions",
              completion_window="24h",
              metadata={
                "description": f"test job for {file}"
              }
          )
          batch_ids.append(batch.id)

        except RateLimitError as e:
          print(f"RateLimitError Error for {file}: {str(e)}")
        except APIError as e:
            print(f"OpenAI Error for {file}: {str(e)}")
        except Exception as ex:
            print(f"Unexpected error: {str(ex)}")
        time.sleep(pause)

file_ids_4o = []
batch_ids_4o = []
process_files(full_files, file_ids_4o, batch_ids_4o)

file_ids_4o_mini = []
batch_ids_4o_mini = []
process_files(mini_files, file_ids_4o_mini, batch_ids_4o_mini)
````

````python
import json

def content_jsonl(input_file, output_file):
            with open(input_file, 'r') as infile:
                for line in infile:
                    try:
                        json_obj = json.loads(line) 
                        content = json_obj.get('response', {}).get('body', {}).get('choices', [{}])[0].get('message', {}).get('content', '') 
                        with open(output_file, 'a') as outfile:
                            outfile.write(json.dumps(content) + '\n')
                    except json.JSONDecodeError:
                        continue

def retrieve_batches(batch_ids, output_jsonl_file, content_output_file):
    for batch_id in batch_ids:
        batch_info = client.batches.retrieve(batch_id)

        input_file_response = client.files.content(batch_info.output_file_id)

        for line in input_file_response.text.splitlines():
            try:
                json_obj = json.loads(line)  
                with open(output_jsonl_file, 'a') as outfile:
                    outfile.write(json.dumps(json_obj) + '\n')  
            except json.JSONDecodeError:
                continue
    content_jsonl(output_jsonl_file, content_output_file)

retrieve_batches(batch_ids_4o_mini, 'response_test_4o_mini.jsonl', 'content_test_4o_mini.jsonl')
retrieve_batches(batch_ids_4o, 'response_test_4o.jsonl', 'content_test_4o.jsonl')
````

````python
import json
import re

def clean_json(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line_number, line in enumerate(infile, start=1):
            line = line.strip()
            
            if not line:
                continue
            
            if line.startswith('"```json\\n') and line.endswith('```"'):
                line = line[10:-3].strip()
            elif line.startswith('"```json\\n') and line.endswith('``` "'):
                line = line[10:-4].strip()
            
            line = line.replace('\\"', '"').replace('\\n', '')[:-1]

            if line.startswith('"'):
                line = line[1:]
            
            line = re.sub(r'\s+(?=(?:[^"]|"[^"]*")*$)', '', line)
            
            try:
                json_obj = json.loads(line)
                outfile.write(json.dumps(json_obj) + '\n')
            except json.JSONDecodeError as e:
                print(e)

# Usage
clean_json('content_test_4o_mini.jsonl', 'cleaned_content_test_4o_mini.jsonl')
clean_json('content_test_4o.jsonl', 'cleaned_content_test_4o.jsonl')
````
